```python
def load_sample():
    """Return the entire text of Hamlet in a string."""
    with open('hamlet.txt') as f:
        sample = f.read()
    return sample

#takes in a string
def delete_space(text):
    txt=list(text)
    for i in range(0, len(text)):
        if (txt[i].isspace()==True or txt[i]=="-" or txt[i]=="/n"):
            txt[i]=' '
    return ''.join(txt)

def delete_punct(text):
    txt=list(text)
    for i in range(0, len(text)):
        if (txt[i].isalpha()==False):
            txt[i]=' '
    return ''.join(txt)
def cleanUp(lst):
    lst1=[]
    for i in range(0, len(lst)):
        if(len(lst[i])>0):
            lst1.append(lst[i])
    return lst1

def parse_sample(text, words_only, sort_list):
    text = delete_space(text)
    if (words_only):
        text=text.lower()
        text=delete_punct(text)
    lst = text.split(' ')
    if(sort_list):
        lst.sort()
    lst = cleanUp(lst)
    return lst


s= load_sample()
parse_sample(s, False, False)
```

```python
parse_sample(s, False, True)
```

```python
import numpy as np
from collections import Counter
def count_freq(lst):
    counts = Counter(lst)
    return counts
    
    """Return a dict of word frequencies given a list of words.
    
    Arguments:
    -- lst: a list of strings
    
    Returns:
    -- freq_dict: a dictionary of word: frequency pairs. The keys are strings
       containing the words, and the values are integers indicating how many
       times each word appears.
    
    Example:
    >>> count_freq(['time', 'after', 'time'])
    {'time': 2, 'after': 1}
    """
```

```python
def freqs(counts):
    # Returns a list of frequencies of each word in dictionary
    return counts.values()

def mean(num):
    """Return the arithmetic mean of a list of numbers."""
    return np.sum(num)/len(num)

def stdev(num):
    """Return the standard deviation of a list of numbers."""
    return np.std(num)
def median(num):
    """Return the median of a list of numbers."""
    num.sort()
    if (len(num)%2!=0):
        return num[len(num)/2]
    if (len(num)%2==0):
        
        return mean([num[(len(num)/2)-1], num[len(num)/2]])
```

## Part 3

```python
def print_stats(lst):
    cf= freqs(count_freq(lst))
    cf = [float(x) for x in cf]
    print "Mean: "+str(mean(cf))
    print "Standard Deviation: "+str(stdev(cf))
    print "Median: "+str(median(cf))
```
